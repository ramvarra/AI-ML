```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.datasets import load_boston
from sklearn.externals import joblib
import sklearn.metrics as sklm
%matplotlib inline

from sklearn.datasets import make_friedman1

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
```

### Polynomial fit

```python
def make_poly_data(f):
    x = np.linspace(0, np.pi * 3, 100)
    #y = np.sin(x) * 1 + np.random.randn(len(x)) * .1 + .1 * x
    y = 2 * x + (np.sin(x) * 2) + np.random.randn(len(x))
    df = pd.DataFrame({'AdSpend': x, 'Sales': y})
    df.to_csv(f, index=None)

def load_poly_data(f, degree=4):
    df = pd.read_csv(f)
    X_scaled = StandardScaler().fit_transform(df.AdSpend.values.reshape(-1, 1))
    poly = PolynomialFeatures(degree=degree)
    X = poly.fit_transform(X_scaled)
    y = df.Sales.values.reshape(-1,1)
    return X, y

def cost(X, y, θ):
    return np.sum(np.square(np.matmul(X, θ) - y))/(2*len(y))

def gradient(X, y, θ):
    e = np.matmul(X, θ) - y
    dJ = np.matmul(X.T, e)/len(y)
    return dJ

def gradient_descent(X, y, lr=0.01, epochs=1000, ϵ=1e-4):
    θ = np.zeros((X.shape[1], 1))
    Jh = [cost(X, y, θ)]
    for epoch in range(epochs):
        dJ = gradient(X, y, θ)
        θ = θ - (lr * dJ)
        Jh.append(cost(X, y, θ))
        v = abs((Jh[-1] - Jh[-2])/Jh[-2])
        if v <= ϵ:
            print(f'Reached min at epoch {epoch}')
            break
        
    return θ, Jh

f = r'DATASETS\AdSpend-vs-Sales.csv'
make_poly_data(f)

fig, ax = plt.subplots(figsize=(16,12))


for degree in [1, 2, 3, 4, 5, 6, 7]:
    
    X, y = load_poly_data(f, degree=degree)
    θ, J_history = gradient_descent(X, y)
    #print(f'θ = {θ} ')
    y_predict = np.matmul(X, θ)
    if degree == 1:
        ax.scatter(X[:,1], y, marker='+', c='r')
    ax.plot(X[:,1], y_predict, label=f'Degree_{degree}')
    r2_score = sklm.r2_score(y, y_predict)
    print(f'Degree = {degree} R2: {r2_score}')

plt.legend()    
```
